# Crear las variables por canton

## Trabajo

```python
import pandas as pd
df["p20"] = df["p_20"].map({"no": 1, "si": 2})
df["p30bi"] = df["p30"].map({"no ha realizado ninguna gestión": 1}).fillna(0).astype(int)
# Generar variable PEA
df["pea"] = ((df["edad"] >= 15) & ((df["p20"] == 2) | ((df["p20"] == 1) & (df["p30bi"] == 0)))).astype(int)
#Genarar PEI
df["pei"] = ((df["edad"] >= 15) & (df["pea"] == 0)).astype(int)
#generar Empleo
df["empleo"] = ((df["pea"]==1) & (df["p20"]==2)).astype(int)
#Generar desempleo o empleo inadecuado
df["desempleo"] = (((df["pea"]==1) & (df["p20"]==1) & (df["p30bi"]==0)) | ((df["pea"]==1) & (df["p20"]==2) & (df["p24"]<40))).astype(int)
#Seguridad social (solo para anio = 2024)
df['p44'] = df['p44'].map({"no": 0, 'si': 1}).fillna(0).astype(int) #se codifica como binaria porque en 2025 es una col de 0s y sirve para hacer un collpase (lamda) filtrado por anio

# Tabular
print(df["pea"].value_counts())
print(df.loc[df["anio"] == 2025, "empleo"].value_counts())
print(df["desempleo"].value_counts())
print(df.loc[df['anio']==2024, 'p44'].value_counts())
```

## Vivienda

```python
#propiedad
import numpy as np
df["propia"] = df["tenencia"].map({"propia": 1}).fillna(0).astype(int)
df["alquiler"] = df["tenencia"].map({"arrendada": 1}).fillna(0).astype(int)
df["otros_tenencia"] = df["tenencia"].map({"otro": 1}).fillna(0).astype(int) 

#hacinamiento
df["per_hab"] = (df['num_per'] / df['num_dorm'])
df['hacinamiento'] = (df['per_hab'] >= 3).astype(int)

#deficit habitacional 
for i in ["est_pds", "est_techo", "est_piso"]:
    df[i] = df[i].map({"bueno": 1, "regular": 0.5, "malo": 0})
df["est_viv"] = (df["est_pds"] + df["est_techo"] + df["est_piso"])/3
df["deficit_hab"] = (df["est_viv"] < 0.5).astype(int)

# Alcantarillado
#usar en el df en lugar de agg_df para los graficos pastel, incluyendo alcantarillado

#Recoleccion de basura
df["basura"] = df["rec_bas"].map({"si": 1}).fillna(0).astype(int)


#modificacion de utlimo momento> mapear urbano y rural a la columna zona 2024
df["zona2"] = df["zona"].map({"no": "rural", "si": "urbana"}).fillna(df["zona"]) #conserva los valores no mapeados, aunque tambien pdoria haberse modificado desde el inicio
#area
df["urbana"] = (df["zona2"] == "urbana").astype(int)
df["rural"] = (df["zona2"] == "rural").astype(int)
```

## Educacion

```python
#ninos y adolescentes que no asisten a la escuela
df["no_asisten"] = df["edbb15_17"].map({"si, no todos asisten": 1}).fillna(0).astype(int)
#num promedio del hogar que asisten a la edu superior 
df['num_ed_sup'] = df['ed_sup'].map({'si': 1}).fillna(0).astype(int)
#mayores 25 sin egb
df["no_egb_25"] = df["edbb25_inc"].map({"si": 1}).fillna(0).astype(int)
#no asistencia edu superior razones economicas
df["no_econ"] = df["razon_ed_sup"].map({"si": 1}).fillna(0).astype(int)


```

# Crear el df agregado por canton

```python
#crear contador 
df['n'] = 1
agg_df = (
    df.groupby("canton").apply(
        lambda x: pd.Series({
            # promedio ponderado de ingreso laboral (ingrl)
            "pob": (x["n"]).sum(),
            "pea": (x["pea"]).sum(),
            # sumas ponderadas 
            "desempleo": (x["desempleo"]).sum(),
            "empleo": (x["empleo"]).sum(),
            "seguro_social": (x.loc[x["anio"]==2024, "p44"]).sum(),
            #Vivienda
            "propia": (x["propia"]).sum(),
            "arrendada": (x["alquiler"]).sum(),
            "otros_tenencia": (x['otros_tenencia']).sum(),
            "hacinamiento": (x['hacinamiento']).sum(),
            "deficit_hab": (x['deficit_hab']).sum(),
            "basura": (x["basura"]).sum(),
            #Educacion 
            "no_asis": (x['no_asisten']).sum(),
            "no_egb25": (x["no_egb_25"]).sum(),
            "no_edsup": (x["no_econ"]).sum(),
            "prom_edsup": (x["num_ed_sup"]).mean(),
            #area
            'urbana' : (x['urbana']).sum(),
            'rural': (x['rural']).sum()
        })
    ).reset_index()
)

##Trabajo
#Crear tasa de desempleo (%)
agg_df["tdesem"] = (agg_df["desempleo"] / agg_df["pea"]) * 100
agg_df["seguro%"] = (agg_df["seguro_social"] / agg_df["empleo"]) * 100


##Vivienda
#Crear tasas de tenecia
agg_df['propia%'] = (agg_df['propia'] / agg_df['pob']) *100
agg_df['arrendada%'] = (agg_df['arrendada'] / agg_df['pob']) *100
agg_df['otros_tenencia%'] =  (agg_df['otros_tenencia'] / agg_df['pob']) *100
#Deficit habitacional
agg_df['deficit_hab%'] = (agg_df['deficit_hab']/agg_df['pob']) *100
#Hacinamiento
agg_df["hacinamiento%"] = (agg_df["hacinamiento"]/agg_df['pob']) *100
# Basura
agg_df['basura%'] = ( agg_df['basura']/ agg_df['pob']) *100

##Educacion
agg_df["no_asis%"] = (agg_df["no_asis"] / agg_df['pob']) *100
agg_df['no_egb25%'] = (agg_df["no_egb25"] / agg_df['pob']) *100
agg_df["no_edsup%"] = (agg_df["no_edsup"] / agg_df['pob']) *100

#area
agg_df["urb"]  = (agg_df["urbana"] / agg_df['pob']) *100
agg_df["rur"] = (100 - agg_df["urb"])

#Visualizar
agg_df.head(15)
```
